# Analysis of the obtained Parameter file
## This notebook contains code to find out the number of frames for each MIDI note in the selected octave. This can be used to choose the number of frames to train on in the network

```python
import pickle
import numpy as np
import matplotlib.pyplot as pyp
```

```python
data = pickle.load(open('./Train_octave_4_Parametric.txt', 'rb'))
```

```python
hist_dict = {}
hist_dict['C'] = [0,0]
hist_dict['C#'] = [0,0]
hist_dict['D'] = [0,0]
hist_dict['D#'] = [0,0]
hist_dict['E'] = [0,0]
hist_dict['F'] = [0,0]
hist_dict['F#'] = [0,0]
hist_dict['G'] = [0,0]
hist_dict['G#'] = [0,0]
hist_dict['A'] = [0,0]
hist_dict['A#'] = [0,0]
hist_dict['B'] = [0,0]
```

```python
el = []
for i in data.keys():
    hist_dict[i.split('_')[0]][0] = hist_dict[i.split('_')[0]][0] + 1
    hist_dict[i.split('_')[0]][1] = hist_dict[i.split('_')[0]][1] + data[i]['cc'].shape[0]
    el.append(data[i]['cc'].shape[0])
```

```python
print('Min Number of frames = ',min(el))
print('Max Number of frames = ',max(el))
```

```python
list_files = []
list_frames = []
for k in hist_dict.keys():
    list_files.append(hist_dict[k][0])
    list_frames.append(hist_dict[k][1])
```

```python
pyp.barh(list(hist_dict.keys()), list_files, color='b');
pyp.xlabel('Number of Recordings')
pyp.ylabel('MIDI(Note)')
pyp.title('Number of Recordings for each Note');
```

```python
pyp.barh(list(hist_dict.keys()), list_frames, color='b');
pyp.xlabel('Number of frames')
pyp.ylabel('MIDI(Note)')
pyp.title('Number of Frames for each Note');
```
